```python
# define the custom ranking for each ordinal group
cut_categories=['Fair','Good','Very Good','Premium','Ideal']
color_categories=['D','E','F','G','H','I','j']
clarity_categories=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
from sklearn.impute import SimpleImputer
```

```python
# above library is used for handle missing value for univarient by (mean,median,mode)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# for feature scaling we used this library
# we need feature scaling for reaching gloabal minima
```

```python
from sklearn.preprocessing import OrdinalEncoder
# this is for performing ordinal encoding
# what we done before with mapping we do it by ordinal encoding
# when ever our categorical feature have rank u use ordinal encoding
```

```python
# for pipeline we import libraries like
# it is connecting simple imputer to standard scaler
from sklearn.pipeline import Pipeline
```

```python
# for grouping all these together we have to import
from sklearn.compose import ColumnTransformer
```

```python
## Numerical Pipeline
# we perform simple imputer or handle missing value
# for get better accuracy we perform scalling
num_pipeline=Pipeline(
    steps=[
    ("imputer",SimpleImputer(strategy="median")),
    ("scaler",StandardScaler())
    ]
)

## Categorical Pipeline
cat_pipeline=Pipeline(
    steps=[
    ('Imputer',SimpleImputer(strategy='most_frequent')),
    ("OrdinalEncoder",OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
    ('scaler',StandardScaler())   
    ]
)
```
